# 코로나 월별 감염자, 예방접종자 수 합 


- (코로나 발생 월)부터 (현재 월 - 1) 월별로 합을 구한다

- 년, 월로 나눈 후 groupby 년도, 월.sum()

- 국내 코로나 확진자-백신접종자 현황_2021년 9월 3주차.xlsx



```python
# 현재 날짜에 맞춰 '2021년  N월 M주차_주요 국가별 코로나 확진자-백신접종자 현황'으로 저장하기
import datetime as dt
from datetime import datetime, timedelta

def get_date(y, m, d):
    s = f'{y:04d}-{m:02d}-{d:02d}'
    return datetime.strptime(s, '%Y-%m-%d')

def get_week_no(y, m, d):
    target = get_date(y, m, d)
    firstday = target.replace(day = 1)
    if firstday.weekday() == 6:
        origin = firstday
    elif firstday.weekday() < 3:
        origin = firstday - timedelta(days=firstday.weekday() + 1)
    else:
        origin = firstday + timedelta(days = 6-firstday.weekday())
    return (target - origin).days // 7 + 1

n = dt.datetime.now()
s = '국내 코로나 확진자-백신접종자 현황_{}년 {}월 {}주차.xlsx'.format(n.year, n.month, get_week_no(n.year, n.month, n.day)) 
```

```python
import pandas as pd
import openpyxl
import xlsxwriter

df = pd.read_excel(s, engine = 'openpyxl')
df
```

```python
is_20210531 = df['날짜'] == '2021-05-31'
is_df = df[is_20210531]

is_df
```

```python
# 변경 api 수정되면 다시 변경
df = df.replace('2021-05-30', '2021-05-31')
```

```python
is_20210531 = df['날짜'] == '2021-05-31'
is_df = df[is_20210531]

is_df
```

```python
df[['년도', '월', '일']] = df['날짜'].str.split('-', n=2, expand=True)
```

```python
# df 복사
df_copy1 = df.copy()
df_copy2 = df.copy()
```

```python
# 날짜 object -> datetime 타입으로 으로 변경
df_copy1['날짜'] = pd.to_datetime(df_copy1['날짜'])
# df의 첫번째 날짜와 마지막 날짜 선택
start_date = df_copy1.iloc[1, 0]
end_date = df_copy1.iloc[len(df)-1, 0]
```

```python
# 매월 마지막 날짜만 추출
tmp = pd.date_range(start_date, end_date, freq='M')
date_df = pd.DataFrame(tmp, columns = ['날짜'])
date_df
```

```python
# date_df['날짜'] == df_copy1['날짜'] 인것만 
df_copy1 = df_copy1[df_copy1['날짜'].isin(date_df['날짜'])]
df_copy1
```

```python
#특정컬럼선택
df_copy2 = pd.DataFrame(df_copy2, columns=['년도', '월', '일', '확진자 수', '1차 백신 접종자 수', '2차 백신 접종자 수']) 

# 년도별, 월별 합계
grouped = df_copy2.groupby([df_copy2['년도'], df_copy2['월']]).sum()
grouped
```

```python
# 멀티 인덱스 -> 열로 변경
grouped.reset_index(inplace=True)
grouped
```

```python
# 두 df의 같은 열을 기준으로 inner join 
merge_inner = pd.merge(grouped, df_flt)
merge_inner
```

```python
#특정컬럼 선택, 열 순서, 이름 변경
result_df = pd.DataFrame(merge_inner, columns=['년도', '월', '일', '확진자 수', '누적 확진자 수', '1차 백신 접종자 수', '누적 1차 백신 접종자 수', '2차 백신 접종자 수', '누적 2차 백신 접종자 수']) 
result_df
```
