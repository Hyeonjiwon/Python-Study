## 공연 시설별 통계

- 년도별로 월별 형식으로 저장 -> 2020-01
- 모든 xls 파일 불러오기 (지역, 공연건수, 개막편수, 상연횟수, 매출액, 예매수 열만)
- 마지막 합계 행만 남겨놓기

- 합계 행만 있는 df 합치기 
- 극장별로 나누기 


```python
import pandas as pd
import openpyxl
import xlsxwriter
```

```python
import os 

file_list = os.listdir()
xls_file_list = [file for file in file_list if file.endswith(".xls")]

print(len(xls_file_list))
print(xls_file_list)
```

```python
result_df = pd.DataFrame()

for f in xls_file_list:
    # 특정 열만 불러오기 
    # 확인
    tmp = pd.read_excel(f, usecols=[0, 2, 4, 6, 8, 10, 12])
    
    # 기간, 규모를 추출하기 위해 행 분리
    tmp1 = tmp.iloc[[1,2], [0]]
    
    period  = str(tmp1.loc[[1]]).lstrip()[22:43]
    size = str(tmp1.loc[[2]]).lstrip()[45:48]
    
    print(period)
    print(size)
    
    tmp2 = tmp.loc[[4, 62]]
    # 헤더 변경 
    df = tmp2.rename(columns=tmp2.iloc[0])
    # 첫번째 행 삭제 
    df = df.drop(df.index[0])
    
    # 매출액은 천원 단위여서 * 1000
    df['매출액'] = df['매출액'] * 1000
    
    # 기간, 규모 열 추가 
    df = df.assign(일자 = period.split('~')[0], 
                   규모 = size)

    # 열 순서 변경 
    col1 = df.columns[:6].to_list()
    col2 = df.columns[-2:].to_list()
    new_col = col2 + col1
    new_df = df[new_col]
    
    # df 합치기
    result_df = pd.concat([result_df, new_df], axis=0, ignore_index=True)

result_df
```

```python
grouped = result_df['규모'].groupby(result_df['규모'])
grouped.size()
```

```python
import xlsxwriter

writer = pd.ExcelWriter('월별 공연매출액_공연시설별_1008.xlsx', engine = 'xlsxwriter')

result_df.to_excel(writer, index = False)

writer.save()
```
