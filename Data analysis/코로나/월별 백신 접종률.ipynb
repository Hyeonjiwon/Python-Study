## 월별 백신 접종률 구하기


- 월별 접종률
- 월별 누적 접종률

```python
import pandas as pd
import openpyxl
import xlsxwriter
```

```python
# 현재 날짜에 맞춰 '2021년  N월 M주차_주요 국가별 코로나 확진자-백신접종자 현황'으로 저장하기
import datetime as dt
from datetime import datetime, timedelta

def get_date(y, m, d):
    s = f'{y:04d}-{m:02d}-{d:02d}'
    return datetime.strptime(s, '%Y-%m-%d')

def get_week_no(y, m, d):
    target = get_date(y, m, d)
    firstday = target.replace(day = 1)
    if firstday.weekday() == 6:
        origin = firstday
    elif firstday.weekday() < 3:
        origin = firstday - timedelta(days=firstday.weekday() + 1)
    else:
        origin = firstday + timedelta(days = 6-firstday.weekday())
    return (target - origin).days // 7 + 1
```

```python
import datetime as dt

n = dt.datetime.now()
s = '국내 코로나 확진자-백신접종자 월별 통계_{}년 {}월.xlsx'.format(n.year, n.month, get_week_no(n.year, n.month-1, n.day))
```

```python
s
```

```python
df = pd.read_excel(s, engine = 'openpyxl')
df = df.fillna(0)
```

```python
df.info()
```

```python
df = df.astype({'년도':'str', '누적 1차 백신 접종자 수':'int64', '누적 2차 백신 접종자 수':'int64'})
df.info()
```

```python
population = 51780579
```

```python
df['1차 백신 접종률'] = df['1차 백신 접종자 수'].apply(lambda e: (e/population)*100)
df['누적 1차 백신 접종률'] = df['누적 1차 백신 접종자 수'].apply(lambda e: (e/population)*100)
df['2차 백신 접종률'] = df['2차 백신 접종자 수'].apply(lambda e: (e/population)*100)
df['누적 2차 백신 접종률'] = df['누적 2차 백신 접종자 수'].apply(lambda e: (e/population)*100)
```

```python
writer = pd.ExcelWriter('국내 코로나 확진자-백신접종자 월별 접종률_{}년 {}월.xlsx'
                        .format(n.year, n.month, get_week_no(n.year, n.month-1, n.day)), 
                        engine = 'xlsxwriter')

df.to_excel(writer, index = False)

writer.save()
```
