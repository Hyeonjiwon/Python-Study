# 코로나 백신 예방접종 현황 API

- 공공데이터활용지원센터
- https://www.data.go.kr/tcs/dss/selectApiDataDetailView.do?publicDataPk=15077756#tab_layer_detail_function

- 보건복지부 접종자 수 질병관리청 접종자 수 비교해서 같으면 사용하기 

- 날짜 / 1차 백신 접종자 수 / 누적 1차 백신 접종자 수 / 2차 백신 접종자 수 / 누적 2차 백신 접종자 수

- baseDate	통계 기준일자

- sido	지역명칭

- firstCnt	당일 통계(1차 접종)

- secondCnt	당일 통계(2차 접종)

- totalFirstCnt	전체 누적 통계(1차 접종)

- totalSecondCnt	전체 누적 통계(2차 접종)

- accumulatedFirstCnt	    전일까지의 누적 통계 (1차 접종)

- accumulatedSecondCnt	전일까지의 누적 통계 (2차 접종)

```python
# 한국의 백신 접종자 수 분석
# https://gunn.kim/post/2020-06-13-seoul-metro-data-openapi/
import urllib.request
import json
import pandas as pd 


perPage = '10000'

url = 'https://api.odcloud.kr/api/15077756/v1/vaccine-stat?serviceKey=SAJ%2FFc5PaHo%2FvmwRQl7j9ul5jp43QfUVR9xuEjZMqoL0B6W%2B%2F%2FRR5JDSY1JN%2BDBIhTDheUhdhBZqR%2FGVOQP%2Btg%3D%3D&page=1&perPage=' + perPage 

response = urllib.request.urlopen(url)
json_str = response.read().decode('utf-8')
json_object = json.loads(json_str)

df_vac = pd.json_normalize(json_object['data'])

is_korea = df_vac['sido'] == '전국'
df_vac_1 = df_vac[is_korea]
```

```python
# totalFirstCnt(전체 누적 통계(1차 접종)) = accumulatedFirstCnt(전일까지의 누적 통계 (1차 접종)) + firstCnt(당일 통계(1차 접종)) 인데
# 왜 당일의 accumulatedFirstCnt != 전일의 totalFirstCnt가 같지 않을까 ... 
31710830 - 31709767
```

```python
# 당일 누적 1차 백신 접종자 수를 당일 일일 1차 백신 접종자 수로 변경
df_vac_1['daily_FirstCnt'] = df_vac_1['totalFirstCnt'].diff()

# 당일 누적 2차 백신 접종자당일 일일 2차 백신 접종자 수로 변경
df_vac_1['daily_SecondCnt'] = df_vac_1['totalSecondCnt'].diff()

# '1차 백신 접종자 수','누적 1차 백신 접종자 수','2차 백신 접종자 수','누적 2차 백신 접종자 수'
df_vac_2 = df_vac_1[['baseDate','daily_FirstCnt','totalFirstCnt','daily_SecondCnt','totalSecondCnt']]

# 날짜 형식 통일
import datetime

for i in range(len(df_vac_2)):
    df_vac_2.iloc[i,0] = datetime.datetime.strptime(df_vac_2.iloc[i,0], '%Y-%m-%d %H:%M:%S').strftime('%Y%m%d')

df_vac_2.columns = ['날짜','1차 백신 접종자 수','누적 1차 백신 접종자 수','2차 백신 접종자 수','누적 2차 백신 접종자 수']
df_vac_2

# 백신접종시작일인 21년 2월 27일 데이터부터 3월 11일까지의 데이터가 생략되어 있음. 후에 추가로 생성 예정
```

```python
import pandas as pd
from pandas import DataFrame
import datetime

# 대한민국 데이터 확진자수와 백신접종자 수 날짜를 기준으로 JOIN연산
data_kor = pd.merge(df_conf_2, df_vac_3, how = 'left',on='날짜')

for i in range(len(data_kor)):
    data_kor.iloc[i,0] = datetime.datetime.strptime(data_kor.iloc[i,0],'%Y%m%d').strftime('%Y-%m-%d')

data_kor
```

```python
# 현재 날짜에 맞춰 '2021년  N월 M주차_국내 코로나 확진자-백신접종자 현황'으로 저장하기
import datetime as dt
from datetime import datetime, timedelta

def get_date(y, m, d):
    s = f'{y:04d}-{m:02d}-{d:02d}'
    return datetime.strptime(s, '%Y-%m-%d')

def get_week_no(y, m, d):
    target = get_date(y, m, d)
    firstday = target.replace(day = 1)
    if firstday.weekday() == 6:
        origin = firstday
    elif firstday.weekday() < 3:
        origin = firstday - timedelta(days=firstday.weekday() + 1)
    else:
        origin = firstday + timedelta(days = 6-firstday.weekday())
    return (target - origin).days // 7 + 1

# wlsxwriter 라이브러리를 사용해 데이터를 시트별로 저장
import xlsxwriter
import datetime as dt

n = dt.datetime.now()
writer = pd.ExcelWriter('국내 코로나 확진자-백신접종자 현황_{}년 {}월 {}주차.xlsx'
                        .format(n.year, n.month, get_week_no(n.year, n.month, n.day)), 
                        engine = 'xlsxwriter')

data_kor.to_excel(writer, index = False, sheet_name = '대한민국')

writer.save()
```
