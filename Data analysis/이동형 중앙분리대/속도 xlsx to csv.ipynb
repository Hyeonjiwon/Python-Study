# xlsx to csv

- 1년치 일자별 속도
- .xlxs 파일 시트별로 불러오기
- 한 df로 합치기
- .csv 파일로 저장

```python
import pandas as pd
import openpyxl
```

```python
import os
```

```python
xlsx_file_list = []

for f in os.listdir():
    if ("_일별_속도" in f) and (".xlsx" in f):
        xlsx_file_list.append(f)
        
xlsx_file_list.sort()  
xlsx_file_list
```

```python
wb = []
sheetList = []

# openpyxl를 이용하여 시트명 가져오기
for i in range(len(xlsx_file_list)):
    wb.append(openpyxl.load_workbook(xlsx_file_list[i]))
    sheetList.append(wb[i].sheetnames)
```

```python
print(len(xlsx_file_list))
print(len(sheetList))
```

```python
csv_file_list = []

for file in xlsx_file_list:
    file_name = file.replace('.xlsx', '.csv')
    csv_file_list.append(file_name)
    
csv_file_list
```

```python
# 엑셀 파일 불러오기
for i in range(len(xlsx_file_list)):
    print(xlsx_file_list[i])
        
    # pandas를 이용하여 각 시트별 데이터 가져오기
    xlsx = pd.ExcelFile(xlsx_file_list[i])
    
    # 빈 데이터 프레임 생성
    vel_df = pd.DataFrame()

    # 시트 불러오기
    for j in sheetList[i]:
        print('%s Sheet의 데이터 입니다.' %j)
        
        df = pd.read_excel(xlsx, j)

        vel_df = vel_df.append(df,ignore_index=True)
        
        # 요일 유일한 값 찾기
        print(df['요일'].unique())
        
        print(len(df))
        print(len(vel_df))

        print('*' * 50)
    
    vel_df.to_csv(csv_file_list[i], index=False, encoding='cp949') 
```
